# Техническое задание

## Лабораторная работа №1

**Тема:** реализация работы с конструктивными действительными числами и применение двух алгоритмов глобальной оптимизации к функции-черному ящику на примере функции Розенброка

## 1. Цель работы

Необходимо реализовать программный модуль для работы с **конструктивными действительными числами**, а также реализовать **два алгоритма оптимизации** вычислимой детерминированной функции, рассматриваемой как **black box**. После этого требуется провести эксперименты и тестирование на **двумерной** и **многомерной** версии **функции Розенброка**.

---

## 2. Исправление и уточнение исходной формулировки

### 2.1. Что в исходном условии было ошибочным

Исходное упоминание **конструктивных комплексных чисел** следует считать ошибкой формулировки.
Для данной лабораторной работы требуется использовать **конструктивные действительные числа**, а не комплексные.

### 2.2. Как правильно понимать объект оптимизации

Целевая функция возвращает значение, которое интерпретируется как **конструктивное действительное число**.
Именно с такими числами должны работать алгоритмы сравнения, выбора лучшего решения, поиска минимума или максимума.

### 2.3. Как правильно понимать функцию Розенброка

Подразумевается **обычная действительная функция Розенброка**.
Нужно поддержать:

* двумерный случай;
* многомерный случай для заранее выбранной размерности `N`.

---

## 3. Общая постановка задачи

Требуется разработать программу, состоящую из двух основных частей:

### 3.1. Модуль конструктивных действительных чисел

Нужно реализовать тип данных, описывающий **конструктивное действительное число** и позволяющий выполнять над ним базовые операции, необходимые для задач оптимизации.

### 3.2. Модуль оптимизации

Нужно реализовать **два метода оптимизации** из лекции, работающих с функцией как с **черным ящиком**, то есть алгоритм не должен зависеть от аналитического вида функции и использовать её внутреннюю формулу напрямую.

---

## 4. Обязательные требования к реализации

## 4.1. Тип данных: конструктивное действительное число

Необходимо реализовать абстракцию конструктивного действительного числа. Минимально требуется:

* представление числа в конструктивном виде;
* возможность получения приближения с заданной точностью;
* операции сравнения;
* базовые арифметические операции, если они нужны выбранной архитектуре решения;
* возможность использовать такие числа как значения целевой функции.

### 4.1.1. Что важно для сравнения

Поскольку работа ведется с конструктивными числами, сравнение может быть не всегда мгновенно однозначным.
Если сравнение основано на приближениях или интервалах, необходимо предусмотреть обработку случаев неопределенности, когда текущей точности недостаточно для надежного вывода о порядке.

### 4.1.2. Допустимые упрощения

Использование конструктивных чисел требуется не как отдельная математическая самоцель, а как часть лабораторной.
Поэтому допустимо реализовать тот объем функциональности, который реально нужен для:

* вычисления значения функции;
* сравнения результатов;
* выбора лучшей точки в алгоритмах оптимизации.

---

## 4.2. Целевая функция как black box

Алгоритмы должны уметь работать с функцией в формате **черного ящика**:

* на вход подается точка пространства;
* на выходе возвращается значение функции;
* аналитическая формула функции не должна быть зашита внутрь алгоритма.

Иными словами, алгоритм должен быть написан так, чтобы его можно было применять не только к функции Розенброка, но и к любой другой детерминированной вычислимой функции того же интерфейса.

### 4.2.1. Ограничения

Не предполагается, что функция:

* обязательно дифференцируема;
* обязательно непрерывна;
* обязательно известна аналитически внутри алгоритма.

Поэтому опора на градиентные методы не является обязательной и даже нежелательна как основа решения.

---

## 4.3. Алгоритмы оптимизации

Необходимо реализовать **два алгоритма оптимизации** из числа рассмотренных на лекции.
По имеющимся уточнениям подходят, например:

* **метод имитации отжига**;
* **генетический алгоритм**;
* **метод шаров** / простой перебор / поиск по покрытию области шарами — если именно этот вариант действительно обсуждался на лекции и принимается преподавателем.

### 4.3.1. Минимальное требование

Достаточно реализовать **любые два алгоритма** из разрешенного лекционного набора.

### 4.3.2. Рекомендуемый выбор

Наиболее разумный и универсальный вариант для этой лабораторной:

1. **имитация отжига**;
2. **генетический алгоритм**.

Это хорошие black-box методы, не требующие знания аналитической структуры функции.

---

## 5. Требования к тестовой функции

## 5.1. Двумерная функция Розенброка

Нужно провести тестирование алгоритмов на классической двумерной функции Розенброка:

[
f(x, y) = (1 - x)^2 + 100(y - x^2)^2
]

### Что нужно показать:

* алгоритм способен искать минимум;
* корректно работает на тестовой функции;
* возвращает точку, близкую к глобальному минимуму;
* возвращает значение функции, близкое к нулю.

---

## 5.2. Многомерная функция Розенброка

Также нужно поддержать многомерный вариант функции Розенброка для выбранной размерности `N`:

[
f(x_1, \dots, x_n) = \sum_{i=1}^{n-1} \left[(1 - x_i)^2 + 100(x_{i+1} - x_i^2)^2\right]
]

### Что нужно показать:

* алгоритм работает не только в 2D;
* алгоритм не зависит от фиксированной размерности;
* реализация корректно принимает `N` как параметр.

---

## 8. Что является обязательным, а что нет

## Обязательно

* заменить комплексные числа на действительные;
* реализовать конструктивные действительные числа;
* реализовать два алгоритма оптимизации;
* использовать black-box постановку;
* протестировать на 2D и многомерной функции Розенброка.

## Желательно, но не обязательно

* численное дифференцирование;
* символьное дифференцирование;
* расширенная политика разрешения неопределенности при сравнении;
* дополнительные тестовые функции помимо Розенброка;
* сравнение производительности на разных размерностях.

## Не требуется

* обязательная работа с комплексными числами;
* обязательное аналитическое знание функции внутри алгоритма;
* обязательная дифференцируемость функции;
* обязательное использование градиентных методов.

## 9. Формальный итог постановки

### Итоговая корректная формулировка лабораторной

Необходимо реализовать:
